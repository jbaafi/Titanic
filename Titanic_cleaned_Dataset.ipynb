```python
import numpy as np
import pandas as pd
```

```python
titanic = pd.read_csv("/Users/jbaafi/Desktop/titanic_data/titanic.csv")
titanic.head()
```

Now we drop the features that are not necessary in the data set eg. name, ticket 

```python
titanic.drop(["Name", "Ticket"], axis = 1, inplace = True)
titanic.head()
```

Create indicator for Cabin

```python
titanic["Cabin_ind"] = np.where(titanic["Cabin"].isnull(), 0, 1)
titanic.head(10)
```

Convert Sex to numeric 

```python
gender_num = {"male": 0, "female": 1}
titanic["Sex"] = titanic["Sex"].map(gender_num)
titanic.head()
```

Drop Cabin and Embarked

```python
titanic.drop(["Cabin", "Embarked"], axis = 1, inplace = True)
titanic.head()
```
